The purpose of this module is to be the main entry point into the CLI-based Video game store management system. This module communicates with a range of other modules in order to carry out the various task of the video game store management system.


```python
import sys
sys.path.append("vsModules")

import gameSearch  as g
import gameRent as r
import gameReturn as ret
import InventoryPruning as ip
from IPython.display import clear_output

def draw_line(length=20, symbol="-"):
    """
    Draws a line consisting of a repeated symbol.

    Parameters:
    length (int): The length of the line. Default is 20.
    symbol (str): The symbol used to draw the line.

    Returns:
    None: This function prints the line and returns nothing.
    """
    line = length * symbol
    print(line)

def display_message(size=10, msg=""):
    """
    Displays a message within a frame of symbols.

    Parameters:
    size (int): The total width of the message display including the frame. Default is 10.
    msg (str): The message to be displayed. Default is an empty string.

    Returns:
    None: This function prints the message within a frame and returns nothing.
    """
    symbol = "^"
    size_msg = len(msg)
    side_size = int((size - size_msg) / 2)
    side_line = side_size * symbol
    draw_line(size)
    print(side_line + msg + side_line)
    draw_line(size)



def display_menu ():
    """
    Displays the main menu options for the user

    Parameters:
    None

    Returns:
    None: This function prints out the menu and returns nothing
    """
    display_message(25, "Main Menu")
    print("1.) Search for Video game")
    print("2.) Rent out Vidoe game")
    print("3.) Return Video game")
    print("4.) Inventory Pruning")
    print("5.) End program")


def main_menu():
    """
    Runs the main menu loop,
    Processes various user inputs

    Parameters:
    None

    Reutrns:
    None: This function calls other functions bases on user input
    and manages overall program flow
    
    """
    while True:
        display_menu()
        option = input("Please enter your option: ")
        clear_output()
        if option == "1" :
            search_term = str(input("Please enter a search term (game title, game ID, platform or genre):"))
            print("SELECT AN OPTION")
            print("1 - Show all games")
            print("2 - Show only available games")
            option = int(input("Option:"))
            results = g.search_games(search_term.lower(),option)
            for lines in results :
                print(lines)
        elif option == "2":
            customer_id = str(input("Enter customer ID:"))
            game_id = str(input("Enter game ID:"))
            r.rent(customer_id, game_id)
        elif option == "3":
            game_id = str(input("Please enter the game ID of the game to be returned:"))
            ret.return_game(game_id)
        elif option == "4" :
            print("SELECT AN OPTION")
            print("1 - View average customer ratings graph")
            print("2 - View rental frequncy graph")
            print("3 - Delete video game from database")
            option = int(input("Option:"))
            if option == 1 :
                ip.avg_ratings()
            elif option == 2 :
                ip.rental_fre()
            elif option == 3 :
                game_id = input("Enter the Game ID of the game you wish to delete:")
                ip.delete_game(game_id)        
        elif option == "5":
            print("Thank you for using Video games store manager, bye.")
            break
        else:
            print("Invalid option")


if __name__=="__main__":
    #Testing the Menu
    main_menu()            
            
```
